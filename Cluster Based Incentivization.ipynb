Problem Statement
Lithion power is the largest producer of electric vehicle(e-vehicle) batteries.
They provide batteries on rent to e-vehicle drivers. Drivers rent a battery
typically for a day and thereafter replacing it with a charged battery from the
company.
Lithion power has a variable pricing model based on the driver's driving
history. Battery life depends on factors such as over speeding, distance
driven per day, etc.
You are supposed to create a cluster model where drivers can be grouped
together based on the driving data and group the datapoints so that drivers
will be incentivized based on the cluster.

```python
# Importing Libraries
# Importing warning module# Instance of rcParams for handling default matplotlib values
# Instance of rcParams for handling default matplotlib values
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (12,6)
```

For the sake of simplicity we will take only two features: Mean distance driven per day & mean % of times, 
a driver drove higher than 5miles per hour over the spped limit


```python
# Reading the data
df = pd.read_csv('driver-data.csv')
df.head()
```

The 'id' column represents the unique id of the driver, 'mean_dist_day' represents mean distance driven by driver per day & 'mean_over_speed_perc' represents the mean percentage of the times that driver drove higher tan five miles per hour over the speed limit

### Data Exploration

```python
# Print data information
df.info()
```

```python
# Print data statistics
df.describe()
```

```python
# Applying KMeans clustering with 2 clusters(minimum number of clusters that can exist in a dataset )
# Drop 'id' column from data
kmeans = KMeans(n_clusters=2)
df_analyze = df.drop('id',axis=1)
```

```python
# Fit KMeans clustering to data
kmeans.fit(df_analyze)
```

It as created the clusters

Using some commands to get some information of these clusters

```python
# Print cluster centers
kmeans.cluster_centers_
```

```python
# Print cluster labels
# Print the number of data points in each cluster
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
# Checking how many unique drivers are there in the first & second cluster
# Print the type of labels
# Count the number of data points in each cluster
print(type(kmeans.labels_)) 
unique,counts = np.unique(kmeans.labels_,return_counts=True)
print(dict(zip(unique,counts)))
```

```python
# Set the theme as white grid, as it is better suited to plots with heavy data elements
# Plot the clusters using the lmplot function from seaborn library such that 'mean_dist_day' on x-axis & 'mean_over_speed_perc' on y-axis
df_analyze['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=df_analyze,hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

We can clearly see from the graph that there are two clusters, one centered around 50 mean distance delay and other around 175.
Also we can see that there are more drivers in the cluster with the delay centerd at  175
Since KMeans clustering gives optimum results when iterated multiple times, lets try out the same with incerasing the number of clusters

```python
# Performing KMeans clustering with 4 clusters
# Get the centroids of the clusters
# Count the number of data points in each cluster
kmeans_4 = KMeans(n_clusters=4)
kmeans_4.fit(df.drop('id',axis=1))
kmeans_4.fit(df.drop('id',axis=1))
print(kmeans_4.cluster_centers_)
unique,counts = np.unique(kmeans_4.labels_,return_counts=True)
kmeans_4.cluster_centers_             
print(dict(zip(unique,counts)))             
```

We can clearly see the difference now in the cluster centers. Also here we have a distribution data points in each cluster

```python
# Add cluster labels to the dataset
# Create a scatter plot to visualize the clusters such that 'mean_dist_day' on x-axis & 'mean_over_speed_perc' on y-axis
df_analyze['cluster'] = kmeans_4.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=df_analyze,hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

Its dense compared to the two cluster plot & hence more optimal. Thus we have clusterd the data with KMeans
